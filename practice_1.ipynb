```python
import nltk
from IPython.display import display
```

```python
grammar = nltk.CFG.fromstring("""
S -> NP VP

NP -> PropN
NP -> N

VP -> Adv VP
VP -> Adv V VP
VP -> V NP
VP -> V NP PP


PP -> prep NP

PropN -> 'Nam' | 'Lan'
N     -> 'thành_phố' | 'đại_học' | 'trường' | 'tin_học' | 'bài' | 'thư_viện'
V     -> 'ở' | 'học' | 'quen' | 'thích'
Adv   -> 'đang' | 'rất' | 'hay'
prep  -> 'ở'

""")
```

```python
parser = nltk.RecursiveDescentParser(grammar)
step_parser = nltk.SteppingRecursiveDescentParser(grammar)
```

```python
data = ['Nam ở thành_phố',
'Nam đang học đại_học',
'Nam quen Lan ở trường',
'Lan rất thích học tin_học',
'Nam hay học bài ở thư_viện']
```

```python
tree = []
for s in data:
    sentence = s.split()
    tree.append(parser.parse(sentence))
    for i in tree[len(tree)-1]:
        display(i)
```

```python
sentence = data[0].split()
step_parser.initialize(sentence)
count = 0
while not step_parser.currently_complete():
    count += 1
    display(step_parser.tree())
    step_parser.step()
print(count)
```
